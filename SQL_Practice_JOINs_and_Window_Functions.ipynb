5. Determining the Highest Salary Increase

```python
sql15 = """
with max_date as (
  select max(from_date) as last_date,
          emp_no
  from salaries
  group by emp_no),
difference_salaries as(
  select s.emp_no,
        salary,
        lag(salary) over (partition by s.emp_no order by from_date) as salary_year_ago,
        (salary - lag(salary) over (partition by s.emp_no order by from_date)) as diff_salary
  from salaries s
  join max_date md on s.emp_no = md.emp_no
  where from_date <= last_date and from_date >= DATE_SUB(last_date, INTERVAL 1 YEAR))
select emp_no,
        salary_year_ago,
        salary,
        diff_salary
from difference_salaries
where diff_salary = (select max(diff_salary) from difference_salaries)
"""
```

I am using last_date, which is the most recent date in each employee's salary history. This is used to select records within the last year for each employee individually

```python
df15 = pd.read_sql(sql15, con)
df15
```

I select all records where the salary date is within the last year from the overall maximum date in the entire table, rather than for each employee individually

```python
sql15_1 = """
with max_date as (
  select max(from_date) as last_date
  from salaries),
year_before as (
    select DATE_SUB((select max(from_date) from salaries), INTERVAL 1 YEAR) as last_year_date),
difference_salaries as (
  select s.emp_no,
        salary,
        lag(salary) over (partition by s.emp_no order by from_date) as salary_year_ago,
        (salary - lag(salary) over (partition by s.emp_no order by from_date)) as diff_salary
  from salaries s
    CROSS JOIN year_before
    where s.from_date <= (select last_date FROM max_date)
    and s.from_date >= (select last_year_date FROM year_before)
)

select emp_no,
      salary_year_ago,
      salary,
      diff_salary
from difference_salaries
where diff_salary = (select max(diff_salary) from difference_salaries)

"""
```

```python
df15_1 = pd.read_sql(sql15_1, con)
df15_1
```

 6. Average Salary and First Hired Employee for Each Job Title

```python
sql16 = """
with first_hire as (
  select first_name,
           last_name,
           title,
           e.hire_date,
           e.emp_no,
  FIRST_VALUE(hire_date) over (partition by title order by hire_date) as first_hire_date
  from employees e
  join titles t on e.emp_no = t.emp_no),
avg_salary as(
  select title,
        avg(salary) as average_salary
  from employees e
  join titles t on e.emp_no = t.emp_no
  join salaries s on e.emp_no = s.emp_no
  group by title
)
select fh.first_name,
       fh.last_name,
       fh.title,
       fh.first_hire_date,
       avs.average_salary
from first_hire fh
join avg_salary avs on fh.title = avs.title
where fh.hire_date = fh.first_hire_date
order by fh.title, fh.first_name, fh.last_name
"""
```

7. Using NTILE to Group Employees by Current Salary
    
    `emp_no` : `419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949`

```python
sql17 = """
with filtered_table as (
  select e.emp_no,
      first_name,
      last_name,
      salary
  from employees e
  join salaries s on e.emp_no = s.emp_no
  where e.emp_no in (419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949)
  and s.to_date = '9999-01-01'
  order by salary desc)
select emp_no,
      first_name,
      last_name,
      salary,
      NTILE(3) over (order by salary desc) as salary_group
from filtered_table
"""
```

```python
df17 = pd.read_sql(sql17, con)
df17
```
